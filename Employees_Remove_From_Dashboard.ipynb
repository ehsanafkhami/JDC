```r
#Employee Table SQL query
"
with table_one as (
select * FROM gdm.gdm_m02_emply_org_da
    where dt=add_months(date_add(sysdate(-1),
         1 - day(sysdate(-1))), -1)
    and leave_dt =""
    )
, table_two as (
select 
oo.erp_acct_no, oo.email, oo.leave_dt, oo.dept_name, o.dept_name as last_month_dept_name,
    case when oo.dept_name=o.dept_name then 1 else 0 end aS FLAG
FROM gdm.gdm_m02_emply_org_da oo
inner join table_one o on o.emply_cd=oo.emply_cd 
where oo.dt=date_add(sysdate(-1),
         1 - day(sysdate(-1)))
and oo.leave_dt =""
    and oo.dept_name is not null
)
select 
erp_acct_no, email, leave_dt, dept_name, dept_name as last_month_dept_nam, "Change Team" as reason, flag
FROM table_two
where flag=0
union
select 
erp_acct_no, email, leave_dt, dept_name, dept_name as last_month_dept_nam, "left JDC" as reason
FROM gdm.gdm_m02_emply_org_da oo
where oo.dt=sysdate(-1)
and leave_dt <>""
"
```

```r
employee=read.csv("Employee.csv")
dashboard=read.csv("Dashboard.csv")
```

```r
 dashboard$emails <- paste(dashboard$send_to, ";",dashboard$cc, ";",dashboard$bcc)
```

```r
dashboard_explode=dashboard %>% 
mutate(emails = strsplit(as.character(emails), ";")) %>% 
unnest(emails)
```

```r
left_employees=employee %>% filter(leave_dt!="") 
```

```r
dashboard_left_employee=inner_join(dashboard_explode, left_employees, by=c("emails"="email"))
```
